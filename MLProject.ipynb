```python
from transformers import AutoTokenizer, AutoModelWithLMHead
```

```python
tokenizer = AutoTokenizer.from_pretrained('t5-base')
model = AutoModelWithLMHead.from_pretrained('t5-base', return_dict=True)
```

```python

from pathlib import Path
 # assign directory
directory = '/content/'
# iterate over files in
# that directory
files = Path(directory).glob('*.txt')
for file in files:
    f=open(file,'r' )
    corpus=f.read()
    f.close()
    inputs = tokenizer.encode("summarize: " + corpus,
                          return_tensors='pt',
                          max_length=512,
                          truncation=True)
    summary_ids = model.generate(inputs, max_length=50, min_length=30, length_penalty=5., num_beams=2)
    summary = tokenizer.decode(summary_ids[0])
    print(file,summary)
    
```
